#Load files 

##new_df is new data including new year's data
##diff_df has old data (no changes or new data for the new year)

```python
import pandas as pd
import numpy as np

def get_data(url): # this function gets data stored as csv file from a given url
  df = pd.read_csv(url)
  return df

url_new = "https://raw.githubusercontent.com/JaySanthanam/Programming-for-data/main/Datasets/LGBF_Data_old.csv"
new_df = get_data(url_new)
url_old = "https://raw.githubusercontent.com/JaySanthanam/Programming-for-data/main/Datasets/LGBF_Data_old_test.csv"
old_df = get_data(url_old)
```

#Identify the size difference between the two dataframes.

```python
def dataframe_difference(df1, df2, which=None):
    """Find rows which are different between two DataFrames."""
    comparison_df = df1.merge(
        df2,
        indicator=True,
        how='outer'
    )
    if which is None:
        diff_df = comparison_df[comparison_df['_merge'] != 'both']  # Identify data changes/updates
    else:
        diff_df = comparison_df[comparison_df['_merge'] == which]
    return diff_df
```

```python
df1 = old_df
df2 = new_df

out_diff = dataframe_difference(df1, df2)
out_diff
```

#Actual code starts here

##1. first find new year's data that is added for this year

```python
#find rows with added data from new year
temp = df2.rename({'Year': 'Year_new'}, axis=1)
merged = temp.merge(df1, how='left', left_on=['LocalAuthority','Code','Year_new'], right_on=['LocalAuthority','Code','Year'])
#merged = temp.merge(df1, how='left', left_on=['LocalAuthority','Year_new'], right_on=['LocalAuthority','Year'],indicator='Match')
new_year_df = np.where(merged['Year'].isnull())[0]
new_year_df
```

# create a new data frame with just the new year's data

```python
#new year added data
new_Data = new_df.loc[new_year_df]
new_Data = new_Data.rename(columns={"Indicator": "New_Indicator", "Numerator": "New_Numerator", "Denominator": "New_Denominator"})
new_Data
```

# Create a dataframe with just changes/updates to old data

```python
#make df with only changes
new_update= new_df.drop(new_df.index[new_year_df], inplace=False).reset_index(drop=True)
new_update.head(15)
```

```python
#compare the old df to find the changes
comparison_values = old_df.values == new_update.values
rows,cols=np.where(comparison_values==False)
rows
```

#Add new columns with the changes (keeping the old values)

```python
change_df = pd.DataFrame()
change_df = old_df.loc[rows]
AddNew_df = new_update.loc[rows]
change_df.insert(loc = 4, column = "New_Indicator", value = AddNew_df["Indicator"].values)
change_df.insert(loc = 6, column = "New_Numerator", value = AddNew_df["Numerator"].values)
change_df.insert(loc = 8, column ="New_Denominator", value = AddNew_df["Denominator"].values)
change_df
```

#Add the new year's data in the end in the new value columns

```python
#Add all changes including new data for the new year

All_changes = pd.concat([change_df,new_Data],  join="outer", axis=0).reset_index()
All_changes
```

#Highlight changes in the new value columns

```python
def set_colormap(x):
    c = f"background-color:red" 
    # DataFrame of styles
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    #set new columns by condition
    columns = len(x.columns)
    for i in range (4,columns-1,2):
        col1 = x.columns[i]
        col2 = x.columns[i+1]
        m = x[col1]!=x[col2]
        # set columns by condition
        df1.loc[m, col2] = c
    return df1

All_changes = All_changes.style.apply(set_colormap, axis=None)
```
