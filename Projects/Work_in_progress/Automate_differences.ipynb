Test code for automated output file

```python
import pandas as pd
import numpy as np

def get_data(url): # this function gets data stored as csv file from a given url
  df = pd.read_csv(url)
  return df

url_new = "https://raw.githubusercontent.com/JaySanthanam/Programming-for-data/main/Datasets/test_new.csv"
new_df = get_data(url_new)
url_old = "https://raw.githubusercontent.com/JaySanthanam/Programming-for-data/main/Datasets/test_old.csv"
old_df = get_data(url_old)
```

```python
old_df.shape[0]
```

```python
old_df.equals(new_df)
```

```python
comparison_values = old_df.values == new_df.values
print(comparison_values)
```

```python
import numpy as np
rows,cols=np.where(comparison_values==False)
```

```python
out_df = get_data(url_old)
```

```python
for item in zip(rows,cols):
    if old_df.iloc[item] != new_df.iloc[item]:
       out_df.iloc[item[0], item[1]] = '{} --> {}'.format(old_df.iloc[item[0], item[1]],new_df.iloc[item[0], item[1]])
```

```python
out_df.shape[0]
```

```python
out_df = out_df.dropna(how='all', axis=0)
out_df.head(10)
```
