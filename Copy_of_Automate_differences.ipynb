```python
def dataframe_difference(df1, df2, which=None):
    """Find rows which are different between two DataFrames."""
    comparison_df = df1.merge(
        df2,
        indicator=True,
        how='outer'
    )
    if which is None:
        diff_df = comparison_df[comparison_df['_merge'] != 'both'] 
    else:
        diff_df = comparison_df[comparison_df['_merge'] == which]
    return diff_df
```

```python
def set_colormap(x):
    c = f"background-color:red" 
    # DataFrame of styles
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    #set new columns by condition
    columns = len(x.columns)
    for i in range (4,columns-1,2):
        col1 = x.columns[i]
        col2 = x.columns[i+1]
        m = x[col1]!=x[col2]
        # set columns by condition
        df1.loc[m, col2] = c
    return df1

All_changes.style.apply(set_colormap, axis=None)
```

```python
def highlight(x):
    c = f"background-color:red" 
    #condition
    m = x['New_LA_Indicator']!=x['LA_IndicatorReal']
    # DataFrame of styles
    df1 = pd.DataFrame('', index=x.index, columns=x.columns)
    # set columns by condition
    df1.loc[m, 'New_LA_Indicator'] = c
    return df1

All_changes.style.apply(highlight, axis=None)
```
